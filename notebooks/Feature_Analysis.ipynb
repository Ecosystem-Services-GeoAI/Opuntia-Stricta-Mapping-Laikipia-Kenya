# Spectral profiles

```python
import matplotlib.pyplot as plt
import pandas as pd

classes = [1, 2, 5, 7, 9]
class_names = ['Tree', 'Grassland', 'Shrubland', 'Bareland', 'Opuntia stricta']
colors = ["#335d2c", "#33a02c", "#eaab1e", "#faff2f", "#e31a1c"]
line_styles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]

band_labels = [
    'Band2(Blue)', 'Band3(Green)', 'Band4(Red)', 'Band5(Red Edge 1)',
    'Band6(Red Edge 2)', 'Band7(Red Edge 3)', 'Band8(NIR)', 'Band8a(Red Edge 4)',
    'Band11(SWIR 1)', 'Band12(SWIR 2)'
]
num_bands = len(band_labels)

month_names = [
    'January', 'February', 'March', 'April', 
    'May', 'June', 'July', 'August', 
    'September', 'October', 'November', 'December'
]

fig, axes = plt.subplots(4, 3, figsize=(14, 18), sharex=True, sharey=True)
axes = axes.flatten()  

for month in range(1, 13):
    ax = axes[month - 1]
    
    band_columns = [f"2023_{month:02d}_Data_band_{i}" for i in range(1, num_bands+1)]
    mean_spectral_by_class = df.groupby('Class')[band_columns].mean()
    
    for cls_idx, cls_value in enumerate(classes):
        y_data = mean_spectral_by_class.loc[cls_value].values
        ax.plot(
            range(num_bands),
            y_data,
            color=colors[cls_idx],
            linestyle=line_styles[cls_idx],
            marker='o',
            markersize=5,
            linewidth=2,
            label=class_names[cls_idx]
        )
    
    ax.set_xticks(range(num_bands))
    ax.set_xticklabels(band_labels, rotation=45, ha='right', fontsize=12)
    ax.tick_params(axis='y', labelsize=12)
    
    ax.set_title(month_names[month - 1], fontsize=18, loc='left')
    
    row = (month - 1) // 3
    col = (month - 1) % 3
    
    if row == 3:
        ax.set_xlabel("Spectral bands", fontsize=18)
    else:
        ax.set_xlabel(None)
    
    if col == 0:
        ax.set_ylabel("Spectral Value", fontsize=18)
    else:
        ax.set_ylabel(None)

handles, labels = axes[-1].get_legend_handles_labels()
fig.legend(
    handles, 
    labels, 
    loc='upper center', 
    ncol=5, 
    bbox_to_anchor=(0.5, 0.97),
    fontsize=18
)

plt.subplots_adjust(
    left=0.06,
    right=0.98,
    top=0.92,
    bottom=0.05,
    wspace=0.0, 
    hspace=0.1
)

plt.savefig("spectral_plot_high_res.jpg", dpi=600, bbox_inches='tight')

plt.show()
```

# Dynamic Time Warping (DTW) distances

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from dtaidistance import dtw
from tslearn.barycenters import dtw_barycenter_averaging

feature_sets = {}
for i in range(1, 11):
    feature_name = f"Band{i}"
    columns_12_months = [
        f"2023_{month:02d}_Data_band_{i}"
        for month in range(1, 13)
    ]
    feature_sets[feature_name] = columns_12_months

feature_sets["gndvi"] = [f"2023_{month:02d}_gndvi_band_1" for month in range(1, 13)]
feature_sets["msavi"] = [f"2023_{month:02d}_msavi_band_1" for month in range(1, 13)]
feature_sets["ndvi"] = [f"2023_{month:02d}_ndvi_band_1" for month in range(1, 13)]
feature_sets["ndwi"] = [f"2023_{month:02d}_ndwi_band_1" for month in range(1, 13)]
feature_sets["precipitation"] = [f"2023_{month:02d}_precipitation_band_1" for month in range(1, 13)]
feature_sets["LST_Night"] = [f"2023_{month:02d}_LST_Night_band_1" for month in range(1, 13)]
feature_sets["LST_Day"] = [f"2023_{month:02d}_LST_Day_band_1" for month in range(1, 13)]
feature_sets["ET"] = [f"2023_{month:02d}_ET_band_1" for month in range(1, 13)]

feature_title_map = {
    "Band1":  "Sentinel-2 Band2",
    "Band2":  "Sentinel-2 Band3",
    "Band3":  "Sentinel-2 Band4",
    "Band4":  "Sentinel-2 Band5",
    "Band5":  "Sentinel-2 Band6",
    "Band6":  "Sentinel-2 Band7",
    "Band7":  "Sentinel-2 Band8",
    "Band8":  "Sentinel-2 Band8a",
    "Band9":  "Sentinel-2 Band11",
    "Band10": "Sentinel-2 Band12",
    "ndvi":         "NDVI",
    "ndwi":         "NDWI",
    "gndvi":        "GNDVI",
    "msavi":        "MSAVI",
    "precipitation": "Precipitation",
    "LST_Night":     "LST Night",
    "LST_Day":       "LST Day",
    "ET":            "Evapotranspiration"
}

category_mapping = {
    1: "Tree",
    2: "Grassland",
    5: "Shrubland",
    7: "Bareland",
    9: "Opuntia stricta"
}

df = df_GSV.copy()

df = df.reset_index(drop=True)

grouped_data = df.groupby('Class')
categories_num = sorted(grouped_data.groups.keys())  # [1,2,5,7,9]
num_categories = len(categories_num)

distance_matrices = {}

for feature_name, columns_12_months in feature_sets.items():
    
    all_time_series = df[columns_12_months].values  # shape: (N, 12)
    scaler = MinMaxScaler()
    all_time_series_normalized = scaler.fit_transform(all_time_series)

    representative_sequences = {}
    for cat in categories_num:
        cat_indices = grouped_data.groups[cat] 
        
        cat_time_series_normalized = all_time_series_normalized[cat_indices]

        # DTW barycenter averaging 
        rep_seq = dtw_barycenter_averaging(cat_time_series_normalized)
        representative_sequences[cat] = rep_seq

    category_distance_matrix = np.zeros((num_categories, num_categories))
    for i in range(num_categories):
        for j in range(i + 1, num_categories):
            dist = dtw.distance(
                representative_sequences[categories_num[i]],
                representative_sequences[categories_num[j]]
            )
            category_distance_matrix[i, j] = dist
            category_distance_matrix[j, i] = dist

    distance_matrices[feature_name] = category_distance_matrix

fig, axes = plt.subplots(6, 3, figsize=(14, 24))
axes = axes.flatten()

feature_list = list(feature_sets.keys())
for idx, feature_name in enumerate(feature_list):
    ax = axes[idx]
    dist_matrix = distance_matrices[feature_name]
    
    category_labels = [category_mapping[c] for c in categories_num]
    sns.heatmap(dist_matrix,
                annot=True,
                cmap='coolwarm',
                square=True,
                xticklabels=category_labels,
                yticklabels=category_labels,
                ax=ax)
    
    ax.set_title(feature_title_map[feature_name], fontsize=16, y=1.0)
    ax.set_xticklabels(ax.get_xticklabels(), fontsize=12, rotation=45)
    ax.set_yticklabels(ax.get_yticklabels(), fontsize=12, rotation=0)

plt.subplots_adjust(
    left=0.05,
    right=0.95,
    top=0.95,
    bottom=0.05,
    wspace=0.6,
    hspace=0.70
)
plt.savefig("dtw_distance_matrices.jpg", dpi=600, bbox_inches='tight')
plt.show()
```
